```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('./Tamil_movies_dataset.csv')
df
```

```python
df[df['Rating'].apply(lambda x:float(x)>=8.5)]
```

```python
df.head()
```

Cleaning Unwanted datas

```python
actors=df['Actor'].value_counts()
lessThan2=actors[actors<2]
lessThan2
```

```python
df['Actor']=df['Actor'].apply(lambda val: 'Others' if val in lessThan2 else val)
df['Actor'].value_counts()
```

Total number of Movies acted by an Actors between 2011 to 2019

```python
plt.figure(figsize=(16,7))
sns.countplot(df['Actor'])
plt.xticks(rotation=90)
```

Average Rating Of an Actors

```python
plt.figure(figsize=(17,4))
sns.barplot(data=df,x='Actor',y='Rating')
plt.xticks(rotation=90)
plt.title('Average Rating Of an Actors')
```

Top 10 Highly Rated Movies

```python
movie=pd.read_csv('./Tamil_movies_dataset.csv')
top_ten=movie.nlargest(10,'Rating')[['MovieName','Actor','Rating','Director']]
top_ten
```

```python
plt.figure(figsize=(10,4))
sns.color_palette("husl", 9)
a=sns.barplot(y=top_ten.MovieName,x='Rating',data=top_ten,hue='Director',dodge=False)
plt.xticks(rotation=90)
sns.move_legend(a,"upper left", bbox_to_anchor=(1.05,1))
plt.title('Top 10 high Rated movies')
```

each movies acted by an actor

```python
a=movie.groupby(['Actor','MovieName'])['MovieName'].count()
a.to_csv('GroupByActor.csv')
```
